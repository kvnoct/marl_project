```python
def tiny_2_plot():
    arr = []
    arr_sum = []
    arr_agent0 = []
    arr_agent1 = []
    for i in range(5):
        tmp0 = np.array(pd.read_csv(f'tiny-2/A2C/agent0/run_{i+1}.csv').Value.tolist())
        tmp1 = np.array(pd.read_csv(f'tiny-2/A2C/agent1/run_{i+1}.csv').Value.tolist())
        arr.append(tmp0-tmp1)
        arr_sum.append(tmp0+tmp1)
        arr_agent0.append(tmp0)
        arr_agent1.append(tmp1)
    a2c_mean_diff = np.mean(arr, axis = 0)
    a2c_std_diff = np.std(arr, axis = 0)
    a2c_mean_sum = np.mean(arr_sum, axis = 0)
    a2c_std_sum = np.std(arr_sum, axis = 0)
    a2c_mean_ag0 = np.mean(arr_agent0, axis = 0)
    a2c_std_ag0 = np.std(arr_agent0, axis = 0)
    a2c_mean_ag1 = np.mean(arr_agent1, axis = 0)
    a2c_std_ag1 = np.std(arr_agent1, axis = 0)
    a2c_x =  pd.read_csv(f'tiny-2/A2C/agent0/run_{i+1}.csv').Step.tolist()
    
    arr = []
    arr_sum = []
    arr_agent0 = []
    arr_agent1 = []
    for i in range(5):
        tmp0 = np.array(pd.read_csv(f'tiny-2/IA2C/agent0/run_{i+1}.csv').Value.tolist())
        tmp1 = np.array(pd.read_csv(f'tiny-2/IA2C/agent1/run_{i+1}.csv').Value.tolist())
        arr.append(tmp0 - tmp1)
        arr_sum.append(tmp0+tmp1)
        arr_agent0.append(tmp0)
        arr_agent1.append(tmp1)
    ia2c_mean_diff = np.mean(arr, axis = 0)
    ia2c_std_diff = np.std(arr, axis = 0)
    ia2c_mean_sum = np.mean(arr_sum, axis = 0)
    ia2c_std_sum = np.std(arr_sum, axis = 0)
    ia2c_mean_ag0 = np.mean(arr_agent0, axis = 0)
    ia2c_std_ag0 = np.std(arr_agent0, axis = 0)
    ia2c_mean_ag1 = np.mean(arr_agent1, axis = 0)
    ia2c_std_ag1 = np.std(arr_agent1, axis = 0)
    ia2c_x =  pd.read_csv(f'tiny-2/IA2C/agent0/run_{i+1}.csv').Step.tolist()
    
    
    fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (14, 6))
    
    
    ax[0].plot(ia2c_x, ia2c_mean_diff)
    ax[0].fill_between(ia2c_x, ia2c_mean_diff-ia2c_std_diff, ia2c_mean_diff+ia2c_std_diff, alpha = 0.2, label = 'IA2C')
    ax[0].plot(ia2c_x, a2c_mean_diff)
    ax[0].fill_between(ia2c_x, a2c_mean_diff-a2c_std_diff, a2c_mean_diff+a2c_std_diff, alpha = 0.2, label = 'A2C')
    ax[0].legend()
    ax[0].set_ylabel("Cumulative Reward Difference")
    ax[0].set_xlabel("Timesteps")

    ax[1].plot(a2c_x, ia2c_mean_sum)
    ax[1].fill_between(a2c_x, ia2c_mean_sum-ia2c_std_sum, ia2c_mean_sum+ia2c_std_sum, alpha = 0.2, label = 'IA2C')
    ax[1].plot(a2c_x, a2c_mean_sum)
    ax[1].fill_between(a2c_x, a2c_mean_sum-a2c_std_sum, a2c_mean_sum+a2c_std_sum, alpha = 0.2, label = 'A2C')
    ax[1].legend()
    ax[1].set_ylabel("Cumulative Reward Sum")
    ax[1].set_xlabel("Timesteps")
    
    plt.tight_layout()
    plt.savefig("tiny_2agents.pdf")
    
    plt.figure()
    
    fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (14, 6))
    
    ax[0].plot(ia2c_x, ia2c_mean_ag0)
    ax[0].fill_between(ia2c_x, ia2c_mean_ag0-ia2c_std_ag0, ia2c_mean_ag0+ia2c_std_ag0, alpha = 0.2, label = 'Agent 1')
    ax[0].plot(ia2c_x, ia2c_mean_ag1)
    ax[0].fill_between(ia2c_x, ia2c_mean_ag1-ia2c_std_ag1, ia2c_mean_ag1+ia2c_std_ag1, alpha = 0.2, label = 'Agent 2')
    ax[0].legend()
    ax[0].set_title("IA2C")
    ax[0].set_ylabel("Cumulative Reward")
    ax[0].set_xlabel("Timesteps")

    ax[1].plot(a2c_x, a2c_mean_ag0)
    ax[1].fill_between(a2c_x, a2c_mean_ag0-a2c_std_ag0, a2c_mean_ag0+a2c_std_ag0, alpha = 0.2, label = 'Agent 1')
    ax[1].plot(a2c_x, a2c_mean_ag1)
    ax[1].fill_between(a2c_x, a2c_mean_ag1-a2c_std_ag1, a2c_mean_ag1+a2c_std_ag1, alpha = 0.2, label = 'Agent 2')
    ax[1].legend()
    ax[1].set_title("A2C")
    ax[1].set_ylabel("Cumulative Reward")
    ax[1].set_xlabel("Timesteps")
    plt.tight_layout()
    plt.savefig("tiny_2agents_agents.pdf")
```

```python
def plot_4_agents():
    arr1 = []
    arr2 = []
    arr3 = []
    arr4 = []

    for i in range(5):
        tmp1 = pd.read_csv(f'tiny-4/agent0/A2C_run_{i+1}.csv').Value.tolist()
        tmp2 = pd.read_csv(f'tiny-4/agent1/A2C_run_{i+1}.csv').Value.tolist()
        tmp3 = pd.read_csv(f'tiny-4/agent2/A2C_run_{i+1}.csv').Value.tolist()
        tmp4 = pd.read_csv(f'tiny-4/agent3/A2C_run_{i+1}.csv').Value.tolist()
        arr1.append(tmp1)
        arr2.append(tmp2)
        arr3.append(tmp3)
        arr4.append(tmp4)

    a2c_mean_ag1 = np.mean(arr1, axis = 0)[-1]
    a2c_mean_ag2 = np.mean(arr2, axis = 0)[-1]
    a2c_mean_ag3 = np.mean(arr3, axis = 0)[-1]
    a2c_mean_ag4 = np.mean(arr4, axis = 0)[-1]
    a2c_std_ag1 = np.std(arr1, axis = 0)[-1]
    a2c_std_ag2 = np.std(arr2, axis = 0)[-1]
    a2c_std_ag3 = np.std(arr3, axis = 0)[-1]
    a2c_std_ag4 = np.std(arr4, axis = 0)[-1]
    
    agents = ['Agent 1', 'Agent 2', 'Agent 3', 'Agent 4']
    counts = [a2c_mean_ag1, a2c_mean_ag2, a2c_mean_ag3, a2c_mean_ag4]
    std = [a2c_std_ag1, a2c_std_ag2, a2c_std_ag3, a2c_std_ag4]
    plt.bar(agents, counts, yerr = std)
    plt.title("Cumulative rewards @ 1e6 timesteps")
    
    plt.tight_layout()
    plt.savefig("tiny_4agents.pdf")
    
    arr1 = []
    arr2 = []
    arr3 = []
    arr4 = []

    for i in range(5):
        tmp1 = pd.read_csv(f'medium-4/agent0/A2C_run_{i+1}.csv').Value.tolist()
        tmp2 = pd.read_csv(f'medium-4/agent1/A2C_run_{i+1}.csv').Value.tolist()
        tmp3 = pd.read_csv(f'medium-4/agent2/A2C_run_{i+1}.csv').Value.tolist()
        tmp4 = pd.read_csv(f'medium-4/agent3/A2C_run_{i+1}.csv').Value.tolist()
        arr1.append(tmp1)
        arr2.append(tmp2)
        arr3.append(tmp3)
        arr4.append(tmp4)

    a2c_mean_ag1 = np.mean(arr1, axis = 0)[-1]
    a2c_mean_ag2 = np.mean(arr2, axis = 0)[-1]
    a2c_mean_ag3 = np.mean(arr3, axis = 0)[-1]
    a2c_mean_ag4 = np.mean(arr4, axis = 0)[-1]
    a2c_std_ag1 = np.std(arr1, axis = 0)[-1]
    a2c_std_ag2 = np.std(arr2, axis = 0)[-1]
    a2c_std_ag3 = np.std(arr3, axis = 0)[-1]
    a2c_std_ag4 = np.std(arr4, axis = 0)[-1]
    
    plt.figure()
    
    agents = ['Agent 1', 'Agent 2', 'Agent 3', 'Agent 4']
    counts = [a2c_mean_ag1, a2c_mean_ag2, a2c_mean_ag3, a2c_mean_ag4]
    std = [a2c_std_ag1, a2c_std_ag2, a2c_std_ag3, a2c_std_ag4]
    plt.bar(agents, counts, yerr = std)
    plt.title("Cumulative rewards @ 1e6 timesteps")
    
    plt.tight_layout()
    plt.savefig("medium_4agents.pdf")
```
